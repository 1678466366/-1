```python
#coding:utf-8
import numpy as np
import pandas as pd

data = pd.read_excel("数据.xlsx")
# 除去NaN列
df = data[data.columns[0:38]]
df[0:40]
```

```python
male_tall = []
female_tall = []
for i in range(0,488):
    if df.loc[i][1]==1:
        female_tall.append(df.loc[i][3])
    else:
        male_tall.append(df.loc[i][3])

sex = df['性别'].values
tall = df['身高'].values
np.corrcoef(sex,tall)
```

```python
import matplotlib.pyplot as plt
import numpy as np



def adjacent_values(vals, q1, q3):
    upper_adjacent_value = q3 + (q3 - q1) * 1.5
    upper_adjacent_value = np.clip(upper_adjacent_value, q3, vals[-1])

    lower_adjacent_value = q1 - (q3 - q1) * 1.5
    lower_adjacent_value = np.clip(lower_adjacent_value, vals[0], q1)
    return lower_adjacent_value, upper_adjacent_value


def set_axis_style(ax, labels):
    ax.get_xaxis().set_tick_params(direction='out')
    ax.xaxis.set_ticks_position('bottom')
    ax.set_xticks(np.arange(1, len(labels) + 1))
    ax.set_xticklabels(labels)
    ax.set_xlim(0.25, len(labels) + 0.75)
    ax.set_xlabel('性别')


# create test data

data = [male_tall,female_tall]

fig, ax1 = plt.subplots(nrows=1, ncols=1, figsize=(10, 8), sharey=True)

ax1.set_title('不同性别身高比对')
ax1.set_ylabel('身高')
ax1.violinplot(data)


# set style for the axes
labels = ['男', '女']
set_axis_style(ax1, labels)

plt.subplots_adjust(bottom=0.15, wspace=0.05)
plt.show()
```

```python
# 主成分分析
from sklearn.decomposition import PCA

#数据读取与清晰，去除Nan列
data = pd.read_excel("数据.xlsx")
data = data[data.columns[2:38]]


pca = PCA(2)
pca.fit(data)
print(pca.components_)
print(pca.explained_variance_ratio_)
```

```python
import numpy as np
import pandas as pd
from factor_analyzer import FactorAnalyzer
import matplotlib.pyplot as plt
data = pd.read_excel("数据.xlsx")
# 取36个因子的数据
df = data[data.columns[2:38]]

# 两个隐藏变量（男，女）
# 充分性测试(Adequacy Test)
# 在做因子分析之前, 我们需要先做充分性检测, 就是数据集中是否能找到这些factor, 我们可以使用下面的两种方式进行寻找.

# Bartlett's Test
# Kaiser-Meyer-Olkin Test
# Bartlett's test of sphericity 是用来检测观察到的变量之间是否关联, 如果检测结果在统计学上不显著, 就不能采用因子分析.
from factor_analyzer.factor_analyzer import calculate_bartlett_sphericity
chi_square_value,p_value=calculate_bartlett_sphericity(df)
print(chi_square_value, p_value)

from factor_analyzer.factor_analyzer import calculate_kmo
kmo_all,kmo_model=calculate_kmo(df)
print(kmo_model)
#  0.8773267790252693 > 0.6 可以做因子分析
```

```python
# Create factor analysis object and perform factor analysis
fa = FactorAnalyzer(36, rotation=None)
fa.fit(df)
# Check Eigenvalues
ev, v = fa.get_eigenvalues()
```

```python
# Create scree plot using matplotlib
# 主成分分析
plt.scatter(range(1,df.shape[1]+1),ev)
plt.plot(range(1,df.shape[1]+1),ev)
plt.title('Scree Plot')
plt.xlabel('Factors')
plt.ylabel('Eigenvalue')
plt.grid()
plt.show()
# 根据图片可以知道潜在变量在1-5个之间，说明除了明显的男女特征差异，还存在差异不够明显的一些情况
```
